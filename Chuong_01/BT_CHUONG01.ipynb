```python
class HinhChuNhat:
    def __init__(self, dai=0, rong=0):
        self.dai = dai
        self.rong = rong

    def nhap(self):
        self.dai = float(input("Nhập chiều dài: "))
        self.rong = float(input("Nhập chiều rộng: "))

    def chu_vi(self):
        return 2 * (self.dai + self.rong)

    def dien_tich(self):
        return self.dai * self.rong

    def xuat(self):
        print(f"Chiều dài: {self.dai}")
        print(f"Chiều rộng: {self.rong}")
        print(f"Chu vi: {self.chu_vi()}")
        print(f"Diện tích: {self.dien_tich()}")

hcn = HinhChuNhat()
hcn.nhap()
hcn.xuat()
```

```python
class ThiSinh:
    def __init__(self, hoten="", toan=0, ly=0, hoa=0):
        self.hoten = hoten
        self.toan = toan
        self.ly = ly
        self.hoa = hoa

    def nhap(self):
        self.hoten = input("Họ tên thí sinh: ")
        self.toan = float(input("Điểm Toán: "))
        self.ly = float(input("Điểm Lý: "))
        self.hoa = float(input("Điểm Hóa: "))

    def tong_diem(self):
        return self.toan + self.ly + self.hoa

    def xuat(self):
        print(f"{self.hoten:20} Toán:{self.toan}  Lý:{self.ly}  Hóa:{self.hoa}  Tổng:{self.tong_diem()}")

ds = []
n = int(input("Nhập số thí sinh: "))
for i in range(n):
    ts = ThiSinh()
    ts.nhap()
    ds.append(ts)

print("\nDanh sách thí sinh:")
for ts in ds:
    ts.xuat()

diem_chuan = float(input("\nNhập điểm chuẩn: "))
print("\nThí sinh trúng tuyển:")
for ts in ds:
    if ts.tong_diem() >= diem_chuan:
        ts.xuat()
```

```python
class PhanSo:
    def __init__(self, tu=0, mau=1):
        self.tu = tu
        self.mau = mau

    def nhap(self):
        self.tu = int(input("Nhập tử số: "))
        self.mau = int(input("Nhập mẫu số: "))

    def hop_le(self):
        return self.mau != 0

    def xuat(self):
        if self.hop_le():
            print(f"{self.tu}/{self.mau}")
        else:
            print("Phân số không hợp lệ!")

ps = PhanSo()
ps.nhap()
ps.xuat()
```

```python
class Stack:
    def __init__(self, capacity):
        self.data = []
        self.capacity = capacity

    def push(self, item):
        if len(self.data) < self.capacity:
            self.data.append(item)
            return True
        return False

    def pop(self):
        if len(self.data) > 0:
            return self.data.pop()
        return None

    def isEmpty(self):
        return len(self.data) == 0

    def isFull(self):
        return len(self.data) == self.capacity

if __name__ == "__main__":
    print("--- Bài 4: Stack cơ bản ---")
    s = Stack(2)
    print(f"Rỗng: {s.isEmpty()}")
    s.push(10)
    s.push(20)
    print(f"Đầy: {s.isFull()}")
    s.push(30) # Thao tác thất bại vì đầy
    print(f"Phần tử lấy ra: {s.pop()}")
```

```python
class Stack:
    def __init__(self, capacity):
        self.data = []
        self.capacity = capacity

    def push(self, item):
        if len(self.data) < self.capacity:
            self.data.append(item)
            return True
        return False

    def pop(self):
        if len(self.data) > 0:
            return self.data.pop()
        return None
    
    def isEmpty(self):
        return len(self.data) == 0

    # Phương thức mới (Bài 6)
    def print(self):
        if self.isEmpty():
            print("Stack rỗng.")
        else:
            print("Nội dung Stack (từ Đỉnh xuống Đáy):")
            for item in reversed(self.data):
                 print(f"| {item} |")
            print("-----")

if __name__ == "__main__":
    print("--- Bài 6: Thêm phương thức print ---")
    s = Stack(4)
    s.push("A")
    s.push("B")
    s.push("C")
    s.print()
    s.pop()
    s.print()
```

```python
class Date:
    def __init__(self, d=1, m=1, y=2000):
        self.ngay = d
        self.thang = m
        self.nam = y
    def display(self):
        print(f"Ngày: {self.ngay:02d}/{self.thang:02d}/{self.nam}", end="")

class Employee:
    def __init__(self, ho_ten, ngay_sinh, ngay_vao_cong_ty):
        self.ho_ten = ho_ten
        self.ngay_sinh = ngay_sinh
        self.ngay_vao_cong_ty = ngay_vao_cong_ty
        
    def in_thong_tin(self):
        print(f"\n--- Thông tin Nhân viên: {self.ho_ten} ---")
        print("Ngày sinh: ", end="")
        self.ngay_sinh.display()
        print("\nNgày vào công ty: ", end="")
        self.ngay_vao_cong_ty.display()
        print()

if __name__ == "__main__":
    print("\n--- Bài 8: Lớp Employee ---")
    ns = Date(1, 1, 1990)
    nvc = Date(10, 5, 2015)
    
    nv1 = Employee("Nguyễn Văn A", ns, nvc)
    nv1.in_thong_tin()
```

```python
class DaGiac:
    def __init__(self, c1, c2):
        self.c1 = c1
        self.c2 = c2

    def tinh_chu_vi(self):
        return 2 * (self.c1 + self.c2) 

class HinhBinhHanh(DaGiac):
    def __init__(self, day, ben, cao):
        super().__init__(day, ben)
        self.h = cao

    def tinh_dien_tich(self):
        return self.c1 * self.h

class HinhChuNhat(HinhBinhHanh):
    def __init__(self, dai, rong):
        super().__init__(dai, rong, rong)

    def in_thong_tin(self):
        print(f"\n--- Hình Chữ Nhật (Dài={self.c1}, Rộng={self.c2}) ---")
        print(f"Chu vi: {self.tinh_chu_vi()}")
        print(f"Diện tích: {self.tinh_dien_tich()}")

if __name__ == "__main__":
    print("\n--- Bài 9: Kế thừa Hình học ---")
    hcn1 = HinhChuNhat(10, 5)
    hcn1.in_thong_tin()
```

```python
import math

class Diem:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

class Elip(Diem):
    def __init__(self, x, y, a, b):
        super().__init__(x, y)
        self.a = a
        self.b = b

    def tinh_dien_tich(self):
        return math.pi * self.a * self.b
    
    def in_thong_tin(self):
        dien_tich = self.tinh_dien_tich()
        print(f"\n--- Elip (a={self.a}, b={self.b}) ---")
        print(f"Diện tích: {dien_tich:.2f}")

class DuongTron(Elip):
    def __init__(self, x, y, r):
        super().__init__(x, y, r, r)

    def in_thong_tin(self):
        dien_tich = self.tinh_dien_tich()
        print(f"\n--- Đường Tròn (Bán kính={self.a}) ---")
        print(f"Diện tích: {dien_tich:.2f}")

if __name__ == "__main__":
    print("\n--- Bài 10: Kế thừa Elip ---")
    elip1 = Elip(0, 0, 5, 3)
    elip1.in_thong_tin()
    
    dt1 = DuongTron(1, 2, 4)
    dt1.in_thong_tin()
```

BÀI 11:

```python
import math

class TamGiac:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
        
    def in_loai(self):
        print(f"Là Tam giác thường (3 cạnh: {self.a}, {self.b}, {self.c})")

class TamGiacVuong(TamGiac):
    def __init__(self, a, b):
        c = math.sqrt(a**2 + b**2)
        super().__init__(a, b, c)

    def tinh_dien_tich(self):
        return 0.5 * self.a * self.b
        
    def in_loai(self):
        dien_tich = self.tinh_dien_tich()
        print(f"Là Tam giác VUÔNG (Cạnh: {self.a}, {self.b}, {self.c}, DT: {dien_tich})")

class TamGiacDeu(TamGiac):
    def __init__(self, canh):
        super().__init__(canh, canh, canh)

    def tinh_dien_tich(self):
        return (self.a**2 * math.sqrt(3)) / 4

    def in_loai(self):
        dien_tich = self.tinh_dien_tich()
        print(f"Là Tam giác ĐỀU (Cạnh: {self.a}, DT: {dien_tich:.2f})")

if __name__ == "__main__":
    print("\n--- Bài 11: Kế thừa Tam giác ---")
    tg1 = TamGiac(3, 4, 5.5)
    tg1.in_loai()
    
    tgv1 = TamGiacVuong(3, 4)
    tgv1.in_loai()
    
    tgd1 = TamGiacDeu(6)
    tgd1.in_loai()
```
