```python
class HinhChuNhat:
    def __init__(self, dai=0, rong=0):
        self.dai = dai
        self.rong = rong

    def nhap(self):
        self.dai = float(input("Nhập chiều dài: "))
        self.rong = float(input("Nhập chiều rộng: "))

    def tinh_chu_vi(self):
        return 2 * (self.dai + self.rong)

    def tinh_dien_tich(self):
        return self.dai * self.rong

    def hien_thi(self):
        print(f"\nThông tin hình chữ nhật:")
        print(f"Chiều dài: {self.dai}")
        print(f"Chiều rộng: {self.rong}")
        print(f"Chu vi: {self.tinh_chu_vi()}")
        print(f"Diện tích: {self.tinh_dien_tich()}")

if __name__ == "__main__":
    hcn = HinhChuNhat()
    hcn.nhap()
    hcn.hien_thi()
```

```python
class ThiSinh:
    def __init__(self, ho_ten="", toan=0, ly=0, hoa=0):
        self.ho_ten = ho_ten
        self.toan = toan
        self.ly = ly
        self.hoa = hoa

    def nhap(self):
        self.ho_ten = input("Nhập họ tên thí sinh: ")
        self.toan = float(input("Nhập điểm Toán: "))
        self.ly = float(input("Nhập điểm Lý: "))
        self.hoa = float(input("Nhập điểm Hóa: "))

    def tinh_tong(self):
        return self.toan + self.ly + self.hoa

    def hien_thi(self):
        print(f"{self.ho_ten:20} | Toán: {self.toan:4.1f} | Lý: {self.ly:4.1f} | Hóa: {self.hoa:4.1f} | Tổng: {self.tinh_tong():5.1f}")

if __name__ == "__main__":
    ds_thi_sinh = []
    n = int(input("Nhập số lượng thí sinh: "))

    for i in range(n):
        print(f"\nNhập thông tin thí sinh thứ {i+1}:")
        ts = ThiSinh()
        ts.nhap()
        ds_thi_sinh.append(ts)

    print("\n--- DANH SÁCH THÍ SINH ---")
    for ts in ds_thi_sinh:
        ts.hien_thi()

    diem_chuan = float(input("\nNhập điểm chuẩn: "))
    print("\n--- DANH SÁCH TRÚNG TUYỂN ---")
    for ts in ds_thi_sinh:
        if ts.tinh_tong() >= diem_chuan:
            ts.hien_thi()
```

```python
class Stack:
    def __init__(self, size):
        self.size = size        
        self.stack = []           

    def push(self, value):
        if not self.isFull():
            self.stack.append(value)
            print(f"Đã thêm {value} vào ngăn xếp.")
        else:
            print("Ngăn xếp đã đầy, không thể thêm!")

    def pop(self):
        if not self.isEmpty():
            value = self.stack.pop()
            print(f"Đã lấy ra {value} khỏi ngăn xếp.")
            return value
        else:
            print("Ngăn xếp rỗng, không thể lấy ra!")

    def isEmpty(self):
        return len(self.stack) == 0

    def isFull(self):
        return len(self.stack) == self.size

    def printStack(self):
        if self.isEmpty():
            print("Ngăn xếp hiện đang rỗng.")
        else:
            print("Nội dung ngăn xếp (từ đáy lên đỉnh):", self.stack)

if __name__ == "__main__":
    n = int(input("Nhập kích thước ngăn xếp: "))
    st = Stack(n)

    while True:
        print("\n=== MENU ===")
        print("1. Thêm phần tử (push)")
        print("2. Lấy phần tử ra (pop)")
        print("3. Kiểm tra rỗng")
        print("4. Kiểm tra đầy")
        print("5. In ngăn xếp")
        print("0. Thoát")

        chon = input("Chọn thao tác: ")

        if chon == "1":
            value = float(input("Nhập giá trị cần thêm: "))
            st.push(value)
        elif chon == "2":
            st.pop()
        elif chon == "3":
            print("Ngăn xếp rỗng?" , st.isEmpty())
        elif chon == "4":
            print("Ngăn xếp đầy?", st.isFull())
        elif chon == "5":
            st.printStack()
        elif chon == "0":
            print("Kết thúc chương trình.")
            break
        else:
            print("Lựa chọn không hợp lệ, vui lòng nhập lại!")
```

```python
class Stack:
    def __init__(self, size):
        self.size = size
        self.stack = []

    def push(self, value):
        if not self.isFull():
            self.stack.append(value)
            print(f"Đã thêm {value} vào ngăn xếp.")
        else:
            print("Ngăn xếp đã đầy!")

    def pop(self):
        if not self.isEmpty():
            value = self.stack.pop()
            print(f"Đã lấy ra {value} khỏi ngăn xếp.")
            return value
        else:
            print("Ngăn xếp rỗng!")

    def isEmpty(self):
        return len(self.stack) == 0

    def isFull(self):
        return len(self.stack) == self.size

    def count(self):
        return len(self.stack)

    def print(self):
        if self.isEmpty():
            print("Ngăn xếp rỗng.")
        else:
            print("Nội dung ngăn xếp (từ đáy lên đỉnh):")
            for i, val in enumerate(self.stack):
                print(f"  [{i}] {val}")

if __name__ == "__main__":
    n = int(input("Nhập kích thước ngăn xếp: "))
    st = Stack(n)

    while True:
        print("\n=== MENU ===")
        print("1. Thêm phần tử (push)")
        print("2. Lấy phần tử ra (pop)")
        print("3. Kiểm tra rỗng")
        print("4. Kiểm tra đầy")
        print("5. Đếm số phần tử (count)")
        print("6. In nội dung ngăn xếp (print)")
        print("0. Thoát")

        chon = input("Chọn thao tác: ")

        if chon == "1":
            value = float(input("Nhập giá trị cần thêm: "))
            st.push(value)
        elif chon == "2":
            st.pop()
        elif chon == "3":
            print("Ngăn xếp rỗng?", st.isEmpty())
        elif chon == "4":
            print("Ngăn xếp đầy?", st.isFull())
        elif chon == "5":
            print(f"Số phần tử hiện có: {st.count()}")
        elif chon == "6":
            st.print()
        elif chon == "0":
            print("Kết thúc chương trình.")
            break
        else:
            print("Lựa chọn không hợp lệ, vui lòng nhập lại!")
```

```python
import datetime

class Date:
    def __init__(self, day=1, month=1, year=2000):
        self.day = day
        self.month = month
        self.year = year

    def display(self):
        print(f"{self.day:02d}/{self.month:02d}/{self.year}")

    def next(self):
        try:
            d = datetime.date(self.year, self.month, self.day)
            d_next = d + datetime.timedelta(days=1)
            self.day, self.month, self.year = d_next.day, d_next.month, d_next.year
        except ValueError:
            print("Lỗi: Ngày không hợp lệ!")

    def input_date(self):
        try:
            self.day = int(input("Nhập ngày: "))
            self.month = int(input("Nhập tháng: "))
            self.year = int(input("Nhập năm: "))
        except:
            print("Dữ liệu không hợp lệ, hãy nhập lại!")

if __name__ == "__main__":
    print("=== NHẬP NGÀY ===")
    d = Date()
    d.input_date()
    print("\nNgày bạn nhập là: ", end="")
    d.display()
    d.next()
    print("Ngày kế tiếp là: ", end="")
    d.display()
```

```python
import datetime

class Date:
    def __init__(self, day=1, month=1, year=2000):
        self.day = day
        self.month = month
        self.year = year

    def display(self):
        print(f"{self.day:02d}/{self.month:02d}/{self.year}", end="")

    def input_date(self):
        try:
            self.day = int(input("  Nhập ngày: "))
            self.month = int(input("  Nhập tháng: "))
            self.year = int(input("  Nhập năm: "))
        except:
            print("Dữ liệu không hợp lệ!")

class Employee:
    def __init__(self):
        self.name = ""
        self.birth = Date()
        self.start = Date()

    def input_employee(self):
        self.name = input("Nhập họ tên nhân viên: ")
        print("Nhập ngày sinh:")
        self.birth.input_date()
        print("Nhập ngày vào công ty:")
        self.start.input_date()

    def display(self):
        print("\n--- Thông tin nhân viên ---")
        print("Họ tên:", self.name)
        print("Ngày sinh: ", end="")
        self.birth.display()
        print("\nNgày vào công ty: ", end="")
        self.start.display()
        print("\n----------------------------")

if __name__ == "__main__":
    n = int(input("Nhập số lượng nhân viên: "))
    employees = []
    for i in range(n):
        print(f"\nNhập thông tin nhân viên thứ {i+1}:")
        e = Employee()
        e.input_employee()
        employees.append(e)

    print("\n===== DANH SÁCH NHÂN VIÊN =====")
    for e in employees:
        e.display()
```

```python
import math

class DaGiac:
    def __init__(self, ds_canh=None):
        self.ds_canh = ds_canh or []

    def nhap(self):
        try:
            n = int(input("Nhập số cạnh của đa giác: "))
            self.ds_canh = []
            for i in range(n):
                val = float(input(f"  Nhập độ dài cạnh thứ {i+1}: "))
                self.ds_canh.append(val)
        except:
            print("Dữ liệu không hợp lệ, nhập lại.")
            self.nhap()

    def chu_vi(self):
        return sum(self.ds_canh)

    def dien_tich(self):
        print("Không xác định diện tích cho đa giác tổng quát. Ghi đè ở lớp con nếu cần.")
        return None

    def hien_thi(self):
        print("Đa giác với các cạnh:", self.ds_canh)
        print("Chu vi:", self.chu_vi())
        area = self.dien_tich()
        if area is not None:
            print("Diện tích:", area)

class HinhBinhHanh(DaGiac):
    def __init__(self, a=0.0, b=0.0, h=0.0):
        super().__init__([a, b, a, b])
        self.a = a
        self.b = b
        self.h = h

    def nhap(self):
        try:
            self.a = float(input("Nhập đáy (a) của hình bình hành: "))
            self.b = float(input("Nhập cạnh bên (b) của hình bình hành: "))
            self.h = float(input("Nhập chiều cao tương ứng với đáy (h): "))
            self.ds_canh = [self.a, self.b, self.a, self.b]
        except:
            print("Dữ liệu không hợp lệ, nhập lại.")
            self.nhap()

    def chu_vi(self):
        return 2 * (self.a + self.b)

    def dien_tich(self):
        return self.a * self.h

    def hien_thi(self):
        print("Hình bình hành:")
        print("  đáy a =", self.a, "cạnh b =", self.b, "chiều cao h =", self.h)
        print("  Chu vi =", self.chu_vi())
        print("  Diện tích =", self.dien_tich())

class HinhChuNhat(HinhBinhHanh):
    def __init__(self, a=0.0, b=0.0):
        super().__init__(a, b, b)
        self.a = a
        self.b = b
        self.h = b
        self.ds_canh = [a, b, a, b]

    def nhap(self):
        try:
            self.a = float(input("Nhập chiều dài (a) của hình chữ nhật: "))
            self.b = float(input("Nhập chiều rộng (b) của hình chữ nhật: "))
            self.h = self.b
            self.ds_canh = [self.a, self.b, self.a, self.b]
        except:
            print("Dữ liệu không hợp lệ, nhập lại.")
            self.nhap()

    def dien_tich(self):
        return self.a * self.b

    def hien_thi(self):
        print("Hình chữ nhật:")
        print("  chiều dài a =", self.a, "chiều rộng b =", self.b)
        print("  Chu vi =", self.chu_vi())
        print("  Diện tích =", self.dien_tich())

class HinhVuong(HinhChuNhat):
    def __init__(self, a=0.0):
        super().__init__(a, a)
        self.a = a
        self.b = a
        self.ds_canh = [a, a, a, a]

    def nhap(self):
        try:
            self.a = float(input("Nhập cạnh a của hình vuông: "))
            self.b = self.a
            self.ds_canh = [self.a, self.a, self.a, self.a]
        except:
            print("Dữ liệu không hợp lệ, nhập lại.")
            self.nhap()

    def hien_thi(self):
        print("Hình vuông:")
        print("  cạnh a =", self.a)
        print("  Chu vi =", self.chu_vi())
        print("  Diện tích =", self.dien_tich())

def menu():
    print("\n--- Bài 9: Quản lý hình ---")
    print("1. Nhập và in Đa giác (tổng quát)")
    print("2. Nhập và in Hình bình hành")
    print("3. Nhập và in Hình chữ nhật")
    print("4. Nhập và in Hình vuông")
    print("0. Thoát")

if __name__ == "__main__":
    while True:
        menu()
        choice = input("Chọn thao tác: ")
        if choice == "1":
            dg = DaGiac()
            dg.nhap()
            dg.hien_thi()
        elif choice == "2":
            hb = HinhBinhHanh()
            hb.nhap()
            hb.hien_thi()
        elif choice == "3":
            hn = HinhChuNhat()
            hn.nhap()
            hn.hien_thi()
        elif choice == "4":
            hv = HinhVuong()
            hv.nhap()
            hv.hien_thi()
        elif choice == "0":
            print("Kết thúc.")
            break
        else:
            print("Lựa chọn không hợp lệ, thử lại.")
```

```python
import math

class Point:
    def __init__(self, x=0.0, y=0.0):
        self.x = x
        self.y = y

    def input_point(self):
        try:
            self.x = float(input("Nhập hoành độ x của điểm tâm: "))
            self.y = float(input("Nhập tung độ y của điểm tâm: "))
        except:
            print("Dữ liệu không hợp lệ, thử lại.")
            self.input_point()

    def display(self):
        print(f"Điểm tâm: ({self.x}, {self.y})", end="")

class Elip(Point):
    def __init__(self, x=0.0, y=0.0, a=1.0, b=1.0):
        super().__init__(x, y)
        self.a = abs(a)
        self.b = abs(b)

    def input_elip(self):
        print("Nhập tâm elip:")
        self.input_point()
        try:
            self.a = float(input("Nhập bán trục lớn a (>0): "))
            self.b = float(input("Nhập bán trục nhỏ b (>0): "))
            if self.a <= 0 or self.b <= 0:
                print("a và b phải lớn hơn 0, nhập lại.")
                self.input_elip()
        except:
            print("Dữ liệu không hợp lệ, thử lại.")
            self.input_elip()

    def area(self):
        return math.pi * self.a * self.b

    def display(self):
        print("Elip:")
        super().display()
        print(f"\n  a = {self.a}")
        print(f"  b = {self.b}")
        print(f"  Diện tích = {self.area():.4f}")

class DuongTron(Elip):
    def __init__(self, x=0.0, y=0.0, r=1.0):
        super().__init__(x, y, r, r)
        self.r = abs(r)

    def input_circle(self):
        print("Nhập tâm đường tròn:")
        self.input_point()
        try:
            self.r = float(input("Nhập bán kính r (>0): "))
            if self.r <= 0:
                print("r phải lớn hơn 0, nhập lại.")
                self.input_circle()
            else:
                self.a = self.b = self.r
        except:
            print("Dữ liệu không hợp lệ, thử lại.")
            self.input_circle()

    def area(self):
        return math.pi * self.r * self.r

    def display(self):
        print("Đường tròn:")
        super(Point, self).display()
        print(f"  Tâm = ({self.x}, {self.y})")
        print(f"  r = {self.r}")
        print(f"  Diện tích = {self.area():.4f}")

def menu():
    print("\n--- BÀI 10: Elip & Đường tròn ---")
    print("1. Nhập và in Elip")
    print("2. Nhập và in Đường tròn")
    print("0. Thoát")

if __name__ == "__main__":
    while True:
        menu()
        choice = input("Chọn thao tác: ")
        if choice == "1":
            e = Elip()
            e.input_elip()
            e.display()
        elif choice == "2":
            c = DuongTron()
            c.input_circle()
            c.display()
        elif choice == "0":
            print("Kết thúc.")
            break
        else:
            print("Lựa chọn không hợp lệ, thử lại.")
```

```python
import math

class TamGiac:
    def __init__(self, a=0.0, b=0.0, c=0.0):
        self.a = float(a)
        self.b = float(b)
        self.c = float(c)

    def nhap(self):
        try:
            self.a = float(input("Nhập cạnh a: "))
            self.b = float(input("Nhập cạnh b: "))
            self.c = float(input("Nhập cạnh c: "))
        except:
            print("Dữ liệu không hợp lệ, hãy nhập lại.")
            self.nhap()

    def hop_le(self):
        return (self.a > 0 and self.b > 0 and self.c > 0 and
                (self.a + self.b > self.c) and
                (self.a + self.c > self.b) and
                (self.b + self.c > self.a))

    def chu_vi(self):
        return self.a + self.b + self.c

    def dien_tich(self):
        p = self.chu_vi() / 2.0
        s = p * (p - self.a) * (p - self.b) * (p - self.c)
        if s < 0:
            return None
        return math.sqrt(s)

    def hien_thi(self):
        print("Tam giác:")
        print("  a =", self.a, " b =", self.b, " c =", self.c)
        if not self.hop_le():
            print("  Tam giác không hợp lệ.")
            return
        print("  Chu vi =", self.chu_vi())
        area = self.dien_tich()
        if area is None:
            print("  Không thể tính diện tích.")
        else:
            print("  Diện tích =", format(area, ".4f"))

class TamGiacVuong(TamGiac):
    def __init__(self, a=0.0, b=0.0, c=0.0):
        super().__init__(a, b, c)

    def nhap(self):
        print("Bạn có thể nhập 2 cạnh vuông góc và tính cạnh huyền, hoặc nhập 3 cạnh.")
        choice = input("Nhập '2' để nhập 2 cạnh vuông góc, khác để nhập 3 cạnh: ")
        if choice == "2":
            try:
                x = float(input("Nhập cạnh thứ nhất (góc vuông): "))
                y = float(input("Nhập cạnh thứ hai (góc vuông): "))
                if x <= 0 or y <= 0:
                    print("Cạnh phải lớn hơn 0, nhập lại.")
                    self.nhap()
                    return
                self.a = x
                self.b = y
                self.c = math.hypot(x, y)
            except:
                print("Dữ liệu không hợp lệ, nhập lại.")
                self.nhap()
        else:
            super().nhap()

    def hop_le(self):
        if not super().hop_le():
            return False
        s1 = self.a * self.a + self.b * self.b
        s2 = self.c * self.c
        eps = 1e-6
        if abs(s1 - s2) < eps:
            return True
        s1 = self.a * self.a + self.c * self.c
        s2 = self.b * self.b
        if abs(s1 - s2) < eps:
            return True
        s1 = self.b * self.b + self.c * self.c
        s2 = self.a * self.a
        if abs(s1 - s2) < eps:
            return True
        return False

    def dien_tich(self):
        if not self.hop_le():
            return None
        if abs(self.a*self.a + self.b*self.b - self.c*self.c) < 1e-6:
            return 0.5 * self.a * self.b
        if abs(self.a*self.a + self.c*self.c - self.b*self.b) < 1e-6:
            return 0.5 * self.a * self.c
        return 0.5 * self.b * self.c

    def hien_thi(self):
        print("Tam giác vuông:")
        super().hien_thi()

class TamGiacCan(TamGiac):
    def __init__(self, a=0.0, b=0.0, c=0.0):
        super().__init__(a, b, c)

    def nhap(self):
        print("Bạn có thể nhập cạnh đáy và cạnh bên (2 lần) hoặc nhập 3 cạnh.")
        choice = input("Nhập '2' để nhập (day, canh), khác để nhập 3 cạnh: ")
        if choice == "2":
            try:
                base = float(input("Nhập đáy (base): "))
                side = float(input("Nhập cạnh bên (side): "))
                if base <= 0 or side <= 0:
                    print("Cạnh phải lớn hơn 0, nhập lại.")
                    self.nhap()
                    return
                self.a = side
                self.b = side
                self.c = base
            except:
                print("Dữ liệu không hợp lệ, nhập lại.")
                self.nhap()
        else:
            super().nhap()

    def hop_le(self):
        if not super().hop_le():
            return False
        if abs(self.a - self.b) < 1e-6 or abs(self.a - self.c) < 1e-6 or abs(self.b - self.c) < 1e-6:
            return True
        return False

    def dien_tich(self):
        if not self.hop_le():
            return None
        return super().dien_tich()

    def hien_thi(self):
        print("Tam giác cân:")
        super().hien_thi()

class TamGiacDeu(TamGiacCan):
    def __init__(self, a=0.0):
        super().__init__(a, a, a)

    def nhap(self):
        try:
            side = float(input("Nhập cạnh của tam giác đều: "))
            if side <= 0:
                print("Cạnh phải lớn hơn 0, nhập lại.")
                self.nhap()
                return
            self.a = self.b = self.c = side
        except:
            print("Dữ liệu không hợp lệ, nhập lại.")
            self.nhap()

    def dien_tich(self):
        if self.a <= 0:
            return None
        return (math.sqrt(3) / 4.0) * self.a * self.a

    def hien_thi(self):
        print("Tam giác đều:")
        super().hien_thi()

def menu():
    print("\n--- BÀI 11: Tam giác ---")
    print("1. Nhập và in Tam giác tổng quát")
    print("2. Nhập và in Tam giác vuông")
    print("3. Nhập và in Tam giác cân")
    print("4. Nhập và in Tam giác đều")
    print("0. Thoát")

if __name__ == "__main__":
    while True:
        menu()
        choice = input("Chọn thao tác: ")
        if choice == "1":
            t = TamGiac()
            t.nhap()
            t.hien_thi()
        elif choice == "2":
            tv = TamGiacVuong()
            tv.nhap()
            tv.hien_thi()
        elif choice == "3":
            tc = TamGiacCan()
            tc.nhap()
            tc.hien_thi()
        elif choice == "4":
            td = TamGiacDeu()
            td.nhap()
            td.hien_thi()
        elif choice == "0":
            print("Kết thúc.")
            break
        else:
            print("Lựa chọn không hợp lệ, thử lại.")
```
