```python
import xml.etree.ElementTree as ET

students = ET.Element("students")
def add_student(id, name, birth, class_name, gender):
    student = ET.SubElement(students, "student")

    ET.SubElement(student, "id").text = id
    ET.SubElement(student, "name").text = name
    ET.SubElement(student, "birth").text = birth
    ET.SubElement(student, "class").text = class_name
    ET.SubElement(student, "gender").text = gender
add_student("SV001", "Nguyễn Văn A", "2006", "18A1", "Nam")
add_student("SV002", "Trần Thị B", "2006", "18A1", "Nữ")
add_student("SV003", "Lê Hoàng C", "2006", "18A1", "Nam")

tree = ET.ElementTree(students)
tree.write("students.xml", encoding="utf-8", xml_declaration=True)

```

```python
import xml.etree.ElementTree as ET

company = ET.Element("company")
def add_staff(id, name, salary):
    staff = ET.SubElement(company, "staff", id=id)
    n = ET.SubElement(staff, "name")
    n.text = name
    s = ET.SubElement(staff, "salary")
    s.text = salary
add_staff("1", "Amar Pandey", "8.5 LPA")
add_staff("2", "Akbhar Khan", "6.5 LPA")
add_staff("3", "Anthony Walter", "3.2 LPA")

tree = ET.ElementTree(company)
ET.indent(tree, space="  ", level=0) 
tree.write("example.xml", encoding="utf-8", xml_declaration=True)




```

```python
from xml.dom import minidom
doc = minidom.parse("example.xml")
company = doc.documentElement
staff_list = company.getElementsByTagName("staff")

print("Company Staff Details:\n")

for s in staff_list:
    staff_id = s.getAttribute("id")
    name = s.getElementsByTagName("name")[0].firstChild.data
    salary = s.getElementsByTagName("salary")[0].firstChild.data

    print(f"ID: {staff_id}")
    print(f"Name: {name}")
    print(f"Salary: {salary}")
    print("----------------------------")
```

```python
from xml.dom import minidom
doc = minidom.parse("example.xml")
company = doc.getElementsByTagName("company")[0]
new_staff = doc.createElement("staff")
new_staff.setAttribute("id", "4")

name = doc.createElement("name")
name_text = doc.createTextNode("David Warner")
name.appendChild(name_text)

salary = doc.createElement("salary")
salary_text = doc.createTextNode("10.2 LPA")
salary.appendChild(salary_text)

new_staff.appendChild(name)
new_staff.appendChild(salary)

company.appendChild(new_staff)

with open("example.xml", "w", encoding="utf-8") as f:
    doc.writexml(f, indent="  ", addindent="  ", newl="\n", encoding="utf-8")
```

```python
import requests
import xml.etree.ElementTree as ET
import csv
import html
import re

url = "http://www.hindustantimes.com/rss/topnews/rssfeed.xml"

response = requests.get(url)
if response.status_code == 200:
    with open("rssfeed.xml", "wb") as f:
        f.write(response.content)
    print("Đã tải và lưu RSS thành 'rssfeed.xml'")
else:
    print("Không tải được RSS feed.")
    exit()

tree = ET.parse("rssfeed.xml")
root = tree.getroot()

items = []
for item in root.findall(".//item"):
    title = item.findtext("title", default="").strip()
    link = item.findtext("link", default="").strip()
    pubDate = item.findtext("pubDate", default="").strip()
    description = item.findtext("description", default="").strip()

    description = html.unescape(description)

    description = re.sub(r"<[^>]*>", "", description)

    description = re.sub(r"[#<>]", "", description)

    items.append({
        "Title": title,
        "Link": link,
        "PubDate": pubDate,
        "Description": description
    })

with open("rssfeed.csv", "w", newline='', encoding="utf-8") as csvfile:
    fieldnames = ["Title", "Link", "PubDate", "Description"]
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(items)

print(f"✅ Đã lưu {len(items)} bài báo vào 'rssfeed.csv'")

```

```python
import json
from collections import Counter, OrderedDict

JSON_FILE = "company.json"

def load_data(path):
    with open(path, "r", encoding="utf-8") as f:
        return json.load(f)

def stats_by_unit(data):
    employees = data.get("employees", [])
    total = len(employees)

    cnt = Counter()
    for e in employees:
        unit = e.get("unit", "Unknown")
        cnt[unit] += 1

    ordered = OrderedDict(sorted(cnt.items(), key=lambda x: x[0]))
    return total, ordered

def print_report(data, total, unit_counts):
    company = data.get("company_name", "(Không có tên công ty)")
    address = data.get("address", "(Không có địa chỉ)")
    print("Kết quả thực hiện chương trình:")
    print(f"Tên công ty: {company}")
    print(f"Địa chỉ: {address}")
    print("----- Thống kê nhân viên theo đơn vị ------")
    i = 1
    for unit, count in unit_counts.items():
        pct = (count / total * 100) if total > 0 else 0
        print(f"{i}. Tên đơn vị: {unit}")
        print(f"   - Số nhân viên: {count}")
        print(f"   - Tỷ lệ: {pct:.2f}%")
        i += 1

def main():
    data = load_data(JSON_FILE)
    total, unit_counts = stats_by_unit(data)
    print_report(data, total, unit_counts)

if __name__ == "__main__":
    main()
```

```python
import json
from datetime import datetime

giao_dich_list = []

print("=== QUẢN LÝ GIAO DỊCH ===")

while True:
    loai = input("Nhập loại giao dịch (Tien/Vang): ")
    so_tien = float(input("Nhập số tiền: "))
    mo_ta = input("Nhập mô tả: ")

    giao_dich = {
        "loai": loai,
        "so_tien": so_tien,
        "mo_ta": mo_ta
    }

    giao_dich_list.append(giao_dich)

    tiep_tuc = input("Tiếp tục nhập giao dịch? (y/n): ")
    if tiep_tuc.lower() != "y":
        break
chon = input("Bạn có muốn ghi dữ liệu vào file JSON không? (1: Có / 0: Không): ")

if chon == "1":
    now = datetime.now()
    ten_file = now.strftime("%Y-%m-%d-%H-%M-%S") + ".json"
    with open(ten_file, "w", encoding="utf-8") as f:
        json.dump(giao_dich_list, f, indent=4, ensure_ascii=False)

    print(f" Dữ liệu đã được lưu vào file: {ten_file}")
else:
    print(" Không ghi dữ liệu vào file.")
```
